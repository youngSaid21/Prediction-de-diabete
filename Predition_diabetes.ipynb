```python
import pandas as pd

# Chargez le jeu de données
data = pd.read_csv("diabetes.csv")

# Sélectionner les personnes atteintes de diabète et ayant une tension artérielle (BloodPressure) supérieure à 70
diabetes_bp_gt_70 = data[(data['Outcome'] == 1) & (data['BloodPressure'] > 70)]

# Compter le nombre de personnes dans ce jeu de données qui répondent aux critères
nombre_personnes = len(diabetes_bp_gt_70)

# Afficher le résultat
print("Le nombre de personnes atteintes du diabète et ayant une tension artérielle supérieure à 70 est :", nombre_personnes)
```

```python
import pandas as pd

# Chargez le jeu de données
data = pd.read_csv("diabetes.csv")

# Filtrer les données pour ne conserver que les lignes où "Outcome" est égal à 1
outcome_1_data = data[data['Outcome'] == 1]

# Calculer la moyenne de la colonne "SkinThickness"
moyenne_skinthickness_outcome_1 = outcome_1_data['SkinThickness'].mean()

# Afficher le résultat
print("La moyenne de 'SkinThickness' lorsque 'Outcome' est égal à 1 est :", moyenne_skinthickness_outcome_1)
```

## Étape 2 : Visualisation des données

```python
import pandas as pd
import matplotlib.pyplot as plt

# Chargez le jeu de données
data = pd.read_csv("diabetes.csv")

# Tracez un histogramme pour chaque colonne
data.hist(figsize=(12, 10))
plt.tight_layout()  # Ajuster la disposition pour éviter les chevauchements
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Chargez le jeu de données
data = pd.read_csv("diabetes.csv")

# Séparer les données en individus diabétiques et non diabétiques
diabetic = data[data['Outcome'] == 1]['Age']
non_diabetic = data[data['Outcome'] == 0]['Age']

# Définir les paramètres de l'histogramme
plt.figure(figsize=(10, 6))

# Tracer les histogrammes
plt.hist(diabetic, bins=20, alpha=0.5, color='red', label='Diabétiques', density=True)
plt.hist(non_diabetic, bins=20, alpha=0.5, color='blue', label='Non diabétiques', density=True)

# Tracer les courbes de densité de probabilité
sns.kdeplot(diabetic, color='red')
sns.kdeplot(non_diabetic, color='blue')

# Ajouter des titres et des légendes
plt.xlabel('Âge')
plt.ylabel('Densité de probabilité')
plt.title('Répartition des âges des individus diabétiques et non diabétiques')
plt.legend()

# Afficher le graphique
plt.grid(True)
plt.show()
```
